# Exploratory Data Analysis - Train Outcomes Dataset
This notebook explores the target variable labels in `train_outcomes.csv`.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')
```

## Loading the Dataset


```python
df = pd.read_csv('../data/train_outcomes.csv')
print(f'Dataset shape: {df.shape}')
df.head()
```

## Data Quality & Missing Values


```python
df.info()
print('\nMissing values per column:')
print(df.isnull().sum())
```

## Duplicate Check


```python
print(f'Duplicate company_id count: {df["company_id"].duplicated().sum()}')
```

## Target Class Distribution (Imbalance Check)


```python
counts = df['defaulted'].value_counts()
percentages = df['defaulted'].value_counts(normalize=True) * 100
for val, count in counts.items():
    print(f'Class {val} (Defaulted = {val}): {count} ({percentages[val]:.2f}%)')

plt.figure(figsize=(6, 4))
sns.countplot(x='defaulted', data=df, palette='Set2')
plt.title('Distribution of Target Variable (defaulted)')
plt.show()
```
